```python
import pandas as pd
import numpy as np

# Sample — assuming df is your DataFrame
# Target: 'Found_fraud'

# Convert 'detection_date' to datetime and then to ordinal
df3['detection_date'] = pd.to_datetime(df3['detection_date'], errors='coerce').apply(lambda x: x.toordinal() if pd.notna(x) else -1)

# Convert 'recovery_number' to numeric, coercing errors
df3['recovery_number'] = pd.to_numeric(df3['recovery_number'], errors='coerce')

# Impute NaN values in 'estimated_daily_energy_consumption' with the mean
df3['estimated_daily_energy_consumption'].fillna(df3['estimated_daily_energy_consumption'].mean(), inplace=True)


X = df3.drop(columns=['Found_fraud'])
y = df3['Found_fraud']

# Convert to numpy
X = X.values
y = y.values
```

```python
# Reshape (samples, time_steps, features)
X = X.reshape((X.shape[0], X.shape[1], 1))
```

```python
 from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.model_selection import StratifiedKFold
from sklearn.metrics import confusion_matrix, classification_report, roc_auc_score, precision_recall_curve, auc
from xgboost import XGBClassifier
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, Dense
from tensorflow.keras.optimizers import Adam
from sklearn.impute import SimpleImputer
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings("ignore")

# Prepare data
X = df3.drop(columns=['Found_fraud'])
y = df3['Found_fraud'].values

# Impute missing values
imputer = SimpleImputer(strategy='mean')
X_imputed = imputer.fit_transform(X)

# Scale features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_imputed)

# KMeans clustering (optional — if you feel clustering adds value)
kmeans = KMeans(n_clusters=5, random_state=42)
cluster_labels = kmeans.fit_predict(X_scaled)

# Add cluster labels as feature
X_clustered = np.concatenate([X_scaled, cluster_labels.reshape(-1, 1)], axis=1)

# Stratified CV setup
cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

# For metrics
all_y_true, all_y_pred, all_y_prob = [], [], []

for fold, (train_idx, val_idx) in enumerate(cv.split(X_clustered, y)):
    print(f"\n▶️ Fold {fold+1}")

    # Split data
    X_train, X_val = X_clustered[train_idx], X_clustered[val_idx]
    y_train, y_val = y[train_idx], y[val_idx]

    # Autoencoder definition
    input_dim = X_train.shape[1]
    input_layer = Input(shape=(input_dim,))
    encoded = Dense(64, activation='relu')(input_layer)
    encoded = Dense(32, activation='relu')(encoded)
    bottleneck = Dense(16, activation='relu')(encoded)
    decoded = Dense(32, activation='relu')(bottleneck)
    decoded = Dense(64, activation='relu')(decoded)
    output_layer = Dense(input_dim, activation='linear')(decoded)

    autoencoder = Model(inputs=input_layer, outputs=output_layer)
    encoder = Model(inputs=input_layer, outputs=bottleneck)

    autoencoder.compile(optimizer=Adam(learning_rate=0.001), loss='mse')

    # Train autoencoder on training fold only
    autoencoder.fit(X_train, X_train,
                    epochs=50, batch_size=32, shuffle=True,
                    validation_split=0.2, verbose=0)

    # Transform data through bottleneck
    X_train_embedded = encoder.predict(X_train)
    X_val_embedded = encoder.predict(X_val)

    # XGBoost classifier
    xgb = XGBClassifier(use_label_encoder=False, eval_metric='logloss', random_state=42)
    xgb.fit(X_train_embedded, y_train)

    # Predict on validation fold
    y_val_pred = xgb.predict(X_val_embedded)
    y_val_prob = xgb.predict_proba(X_val_embedded)[:, 1]

    # Collect results
    all_y_true.extend(y_val)
    all_y_pred.extend(y_val_pred)
    all_y_prob.extend(y_val_prob)

# Final metrics after CV
cm = confusion_matrix(all_y_true, all_y_pred)
tn, fp, fn, tp = cm.ravel()

print("\nFinal Confusion Matrix:\n", cm)
print("\nClassification Report:\n", classification_report(all_y_true, all_y_pred, zero_division=0))
print("\nROC AUC Score:", roc_auc_score(all_y_true, all_y_prob))

# Precision-Recall AUC
precision, recall, _ = precision_recall_curve(all_y_true, all_y_prob)
pr_auc = auc(recall, precision)
print("Precision-Recall AUC:", pr_auc)

# Class-wise accuracy
accuracy_class_0 = tn / (tn + fp)
accuracy_class_1 = tp / (tp + fn)
print(f"\nAccuracy for Class 0: {accuracy_class_0:.4f}")
print(f"Accuracy for Class 1: {accuracy_class_1:.4f}")
```
